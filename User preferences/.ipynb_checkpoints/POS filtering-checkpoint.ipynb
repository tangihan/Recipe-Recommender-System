```python
food_word_list = []
ingredient_synset = wn.synsets("food")

def recursive_hypernym(food_word_list,ingredient_synset,n,m,index):
    if n == 1:
        return 
    elif m == n:
        ingredient_list = [ingredient_name.name().split('.')[0] for ingredient_name in ingredient_synset[index].hyponyms()]
        for word in ingredient_list:

            food_word_list.append(word)
            recursive_hypernym(food_word_list,wn.synsets(word),n-1,m,index)
    else:
        
        for syn in ingredient_synset: 
            ingredient_list = [ingredient_name.name().split('.')[0] for ingredient_name in syn.hyponyms()]
            for word in ingredient_list:

                food_word_list.append(word)
                recursive_hypernym(food_word_list,wn.synsets(word),n-1,m,index)
    return food_word_list

# parameters
# food_word_list = empty list
# synset of word = wn.synset("word")
# n = how many iterations you want
# m = conditional iteration count to exempt the first iteration for the below index
# index = what index of the first synset that you want
food_word_list = recursive_hypernym(food_word_list,ingredient_synset,4,4,1)
print(food_word_list)
```

```python
new_food_word_list = []
for food in food_word_list:
    for food in food.split("_"):
        if food not in new_food_word_list and food in pos_tag
            new_food_word_list.append(food)



df = pd.DataFrame(new_food_word_list)
df.to_csv("new_pos_tag.csv",index=False )

df.head()

```
